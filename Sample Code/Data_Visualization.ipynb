Process to data wrangling:
1. Employee engagement data is in raw form.
2. Data get imported as a DataFrame and later cleaned.
3. Cleaned data is then transformed into graphs, from which findings can be derivevd.
4. We analyze this data to communicate the final results.

```
# import all library in here.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

```
# upload the data through here.
from google.colab import files
files.upload()
```

```
# read the data
data = pd.read_csv("Iris.csv")
data
```

```
# display the data using number of row and number of column
data.iloc[:,1]
```

```
# show data based on condition
data[data['PetalWidthCm'] > 2]
data['PetalWidthCm'] > 2
```

```
# mean, median, variance and standard deviation
print("Mean : " + str(np.mean(data["SepalLengthCm"])))
print("Median : " + str(np.median(data["SepalLengthCm"])))
print("Variance : " + str(np.var(data["SepalLengthCm"])))
print("Standard Deviation : "+ str(np.std(data["SepalLengthCm"])))
```

```
plt.plot(data['SepalLengthCm'], label="SepalLength")
plt.plot(data['SepalWidthCm'], label="SepalWidth")
plt.legend()
plt.figure()
```

```
plt.figure(figsize=(10, 6), dpi=300)
plt.title("Iris Dataset")
sns.barplot(x=data['SepalLengthCm'], y=data['SepalLengthCm'])
plt.ylabel("Sepal Length")
```
